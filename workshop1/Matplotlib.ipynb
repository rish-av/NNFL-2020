# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

Using Matplotlib inline, the plots will be embedded into the notebook as png files

#### Two interfaces

* Matplotlib offers two kinds of interfaces. One is a MATLAB-style interface, and other is an object oriented style interface

https://jakevdp.github.io/PythonDataScienceHandbook/04.00-introduction-to-matplotlib.html

## Matplotlib Interface

```python
# Creating an x-axis for plotting
x = np.linspace(0, 10, 100)
```

```python
# Simple plot
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
# Dashes in the plot
plt.plot(x, np.sin(x), '.')
plt.plot(x, np.cos(x), '--');
```

```python
# Save the plot
fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), '--');
fig.savefig('sample_plot.png')
```

```python
# Creating multiple plots in one figure

# Create the first of two panels and set current axis
plt.subplot(2,1,2) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# Create the second panel and set current axis
plt.subplot(2,1,1)
plt.plot(x, np.cos(x))
```

## Object Oriented Interface

```python
# First create a figure
%matplotlib inline
fig = plt.figure()
```

```python
# Add Subplots to the figure
ax1,ax2,ax3,ax4 = [fig.add_subplot(2,2,i) for i in range(1,5)]
```

```python
ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tan(x))
ax4.plot(x, np.exp(-x))
```
